```python
# -*- coding: utf-8 -*-
# Analisis Sentimen MBG - KNN Full Code

# Import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import warnings

warnings.filterwarnings("ignore")

# =========================
# 1. Muat dan bersihkan dataset
# =========================
data = pd.read_csv("dc_tweets_mbg_fix.csv")

print("ðŸ“Š Jumlah data awal:", len(data))

# Hapus baris dengan nilai kosong (NaN) atau string kosong pada kolom 'text'
data = data.dropna(subset=['text'])
data = data[data['text'].str.strip() != '']

print("âœ… Jumlah data setelah pembersihan:", len(data))

# =========================
# 2. Siapkan fitur dan label
# =========================
X_text = data['text']          # Hanya gunakan teks
y = data['Sentimen_Title'].astype(str)

# =========================
# 3. TF-IDF Vectorization
# =========================
tfidf_vectorizer = TfidfVectorizer(max_features=3000)  # Batasi fitur agar KNN lebih cepat
X_tfidf = tfidf_vectorizer.fit_transform(X_text).toarray()

print("âœ… Dimensi TF-IDF:", X_tfidf.shape)

# =========================
# 4. Split Data: Train & Test
# =========================
X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf, y, test_size=0.1, random_state=42, stratify=y
)

print("ðŸ“š Data train:", X_train.shape)
print("ðŸ§ª Data test:", X_test.shape)

# =========================
# 5. Balancing Data dengan SMOTE pada training set
# =========================
smote = SMOTE(random_state=32)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

print("ðŸ”„ Jumlah sampel train sebelum SMOTE:", X_train.shape[0])
print("âœ… Jumlah sampel train setelah SMOTE:", X_train_res.shape[0])

# Visualisasi distribusi kelas setelah SMOTE
plt.figure(figsize=(6, 4))
y_train_res.value_counts().plot(kind="bar", color=["#2ecc71", "#e74c3c", "#f1c40f"])
plt.title("Distribusi Kelas Training Setelah SMOTE")
plt.xlabel("Kelas Sentimen")
plt.ylabel("Jumlah Sampel")
plt.show()

# =========================
# 6. Latih Model K-Nearest Neighbors
# =========================
model = KNeighborsClassifier(n_neighbors=5, metric='cosine')  # cosine lebih cocok untuk TF-IDF
model.fit(X_train_res, y_train_res)

# =========================
# 7. Prediksi data uji
# =========================
y_pred = model.predict(X_test)

pred_df = pd.DataFrame({
    'Asli': y_test.values,
    'Prediksi': y_pred
})
print("\nðŸ“‹ Contoh Hasil Prediksi:")
print(pred_df.head(10))

# =========================
# 8. Hitung TP, FP, FN, TN per kelas
# =========================
labels = model.classes_
cm = confusion_matrix(y_test, y_pred, labels=labels)

print("\nðŸ“Š TP, FP, FN, TN per kelas:")
print("==========================================")
for i, cls in enumerate(labels):
    TP = cm[i, i]
    FP = cm[:, i].sum() - TP
    FN = cm[i, :].sum() - TP
    TN = cm.sum() - (TP + FP + FN)
    
    print(f"\nKelas: {cls}")
    print(f"  TP = {TP}")
    print(f"  FP = {FP}")
    print(f"  FN = {FN}")
    print(f"  TN = {TN}")
    print("------------------------------------------")

# =========================
# 9. Fungsi untuk menampilkan hasil evaluasi
# =========================
def display_results(y_test, y_pred, cm_title="Confusion Matrix"):
    cm = confusion_matrix(y_test, y_pred, labels=model.classes_)
    
    plt.figure(figsize=(7, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap="Blues",
                xticklabels=model.classes_,
                yticklabels=model.classes_)
    plt.title(cm_title)
    plt.xlabel("Prediksi")
    plt.ylabel("Aktual")
    plt.show()
    
    print("\n=== Classification Report ===")
    print(classification_report(y_test, y_pred))
    print("=============================\n")

display_results(y_test, y_pred)

# =========================
# 10. Hitung akurasi keseluruhan
# =========================
acc = accuracy_score(y_test, y_pred)
print(f"ðŸŽ¯ Accuracy  : {acc:.4f}")
print('-------------------------------------------')
```
