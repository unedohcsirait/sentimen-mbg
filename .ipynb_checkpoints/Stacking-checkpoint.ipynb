```python
# -*- coding: utf-8 -*-
"""Safe Stacking Classifier - tanpa data leakage"""

# Import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# ===============================
# 1ï¸âƒ£ Muat dataset
# ===============================
data = pd.read_csv("dc_tweets_mbg_final.csv")
print("ðŸ“Š Jumlah data awal:", len(data))
data = data.dropna(subset=['text'])
data = data[data['text'].str.strip() != '']
print("âœ… Jumlah data setelah pembersihan:", len(data))

# ===============================
# 2ï¸âƒ£ Siapkan fitur & label
# ===============================
X_text = data['text']  # Hanya teks
y = data['Sentimen_Title'].astype(str)

# ===============================
# 3ï¸âƒ£ Split train/test
# ===============================
X_train_text, X_test_text, y_train, y_test = train_test_split(
    X_text, y, test_size=0.2, random_state=42, stratify=y
)
print("ðŸ“š Data train:", X_train_text.shape)
print("ðŸ§ª Data test :", X_test_text.shape)

# ===============================
# 4ï¸âƒ£ TF-IDF
# ===============================
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train_text).toarray()
X_test_tfidf = tfidf_vectorizer.transform(X_test_text).toarray()
print("âœ… Dimensi TF-IDF train:", X_train_tfidf.shape)
print("âœ… Dimensi TF-IDF test :", X_test_tfidf.shape)

# ===============================
# 5ï¸âƒ£ SMOTE pada train set
# ===============================
smote = SMOTE(random_state=32)
X_train_res, y_train_res = smote.fit_resample(X_train_tfidf, y_train)
print("ðŸ”„ Jumlah sampel train sebelum SMOTE:", X_train_tfidf.shape[0])
print("âœ… Jumlah sampel train setelah SMOTE:", X_train_res.shape[0])

# Visualisasi distribusi kelas setelah SMOTE
plt.figure(figsize=(6, 4))
y_train_res.value_counts().plot(kind="bar", color=["#2ecc71", "#e74c3c", "#f1c40f"])
plt.title("Distribusi Kelas Train Setelah SMOTE")
plt.xlabel("Kelas Sentimen")
plt.ylabel("Jumlah Sampel")
plt.show()

# ===============================
# 6ï¸âƒ£ Base Models Evaluation
# ===============================
base_models = [
    ('KNN', KNeighborsClassifier()),
    ('RandomForest', RandomForestClassifier(n_estimators=100, random_state=42)),
    ('LogisticRegression', LogisticRegression(max_iter=500))
]

print("\n=== Evaluasi Base Models ===")
for name, model in base_models:
    model.fit(X_train_res, y_train_res)
    y_pred_base = model.predict(X_test_tfidf)
    acc = accuracy_score(y_test, y_pred_base)
    print(f"\n{name} Accuracy: {acc:.4f}")
    print(classification_report(y_test, y_pred_base))
    print("-"*50)

# ===============================
# 7ï¸âƒ£ Stacking Classifier
# ===============================
stack_model = StackingClassifier(
    estimators=base_models,
    final_estimator=LogisticRegression(max_iter=500),
    cv=5
)
stack_model.fit(X_train_res, y_train_res)

# Prediksi
y_pred_stack = stack_model.predict(X_test_tfidf)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred_stack, labels=stack_model.classes_)

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap="Blues",
            xticklabels=stack_model.classes_,
            yticklabels=stack_model.classes_)
plt.title("Confusion Matrix - Stacking Classifier")
plt.xlabel("Prediksi")
plt.ylabel("Aktual")
plt.show()

# Classification Report
print("\n=== Classification Report - Stacking ===")
print(classification_report(y_test, y_pred_stack))

# Accuracy
acc_stack = accuracy_score(y_test, y_pred_stack)
print(f"ðŸŽ¯ Accuracy Stacking Classifier: {acc_stack:.4f}")
```
